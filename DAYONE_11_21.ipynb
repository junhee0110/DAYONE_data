```
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os

from sklearn.preprocessing import Normalizer
```

```
dir= os.listdir()
print(dir)
```

```
del dir[3]
print(dir)
```

```
data = []
for x in dir:
  data.append(pd.read_csv(x))
```

```

data
```

```
for x in range(len(data)-1):
  del data[x]['Country']
```

```
data[0].head()
```

```
transform = []
for x in range(len(data)-1):
  transform.append(Normalizer().fit_transform(data[x]))
```

```
transform[0]
```

```
trans_data = []
for x in range(len(data)-1):
  trans_data.append(pd.DataFrame(transform[x],columns = data[x].columns, index = data[x].index))

print(trans_data)
```

```
ans = data[4]['Deaths']
print(ans)
```

```
files = 0
col = 5

print(dir[files])
print(trans_data[files].columns[col])
plt.scatter(trans_data[files].iloc[:,col],ans)
plt.xlabel(trans_data[files].columns[col])
plt.show()
```
