```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras

from sklearn.neighbors import KNeighborsRegressor as knn
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor as rfr
from sklearn.preprocessing import Normalizer
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split as tts
```

```
data = pd.read_csv("Food_Supply_Quantity_kg_Data.csv")
data.head()
```

```
Country = list(data['Country'])
print(Country)
```

```
data = data.iloc[:,1:]
data.head()
```

```
ans = pd.read_csv("Common_Data.csv")
ans.head()
```

```
data = pd.concat([data,ans.loc[:,'Obesity':'Undernourished']],axis=1)
data.head()
```

```
ans = ans['Deaths']
print(ans)
```

```
X_train, X_test, Y_train, Y_test = tts(data, ans, test_size = 0.33, random_state = 4)
print(X_train)
print(X_test)
print(Y_train)
print(Y_test)
```

```
model_R = Pipeline([
                    ('norm', Normalizer()),
                    ('RF',rfr())
])
```

```
model_R.fit(X_train,Y_train)
```

```
print(model_R.score(X_train,Y_train))
print(model_R.score(X_test,Y_test))
```

```
model_S = Pipeline([
                    ('norm', Normalizer()),
                    ('SVR',SVR())
])
```

```
model_S.fit(X_train,Y_train)
```

```
print(model_S.score(X_train,Y_train))
print(model_S.score(X_test,Y_test))
```

```
model_K = Pipeline([
                    ('norm', Normalizer()),
                    ('KNN',knn())
])
```

```
model_K.fit(X_train,Y_train)
```

```
print(model_K.score(X_train,Y_train))
print(model_K.score(X_test,Y_test))
```

```
model_A = keras.Sequential([
    keras.layers.Dense(27, activation='relu'),
    keras.layers.Dense(1024, activation='relu'),
    keras.layers.Dense(1024, activation='relu'),
    keras.layers.Dense(1024, activation='relu'),
    keras.layers.Dense(1024, activation='relu'),
    keras.layers.Dense(5, activation='linear')
])
```

```
model_A.compile(optimizer='adam',
              loss='mean_squared_error',
              metrics=['accuracy'])
```

```
history = model_A.fit(X_train, 
                    Y_train, 
                    batch_size = 20, 
                    epochs=10)
```

```
test_loss, test_acc = model_A.evaluate(X_test,  Y_test, verbose=1)
```

```
print('\n테스트 정확도:', test_acc)
```
